```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

life = pd.read_excel("Life_Expectancy.xlsx")
#type(life)
life.dropna()
life.head(16)
```

```python
#large dataset with many values so a small section of it has to be used 
life["Country"].unique()
```

```python
life['Population']=life['Population']/1000000
life.rename(columns = {'Population':'Population in millions'}, inplace = True)
life.head()
```

```python
countries = life.groupby('Country')
IND =countries.get_group('India')
IND
```

```python
IND_temp=IND.copy()
IND_temp["Life Expectancy"]=(IND_temp["Life Expectancy"]/IND_temp["Life Expectancy"].abs().max())
IND_temp["CO2 emissions"]=(IND_temp["CO2 emissions"]/IND_temp["CO2 emissions"].abs().max())
plt.plot(IND_temp["Life Expectancy"],IND_temp["Year"]) #orange
plt.plot(IND_temp["CO2 emissions"],IND["Year"]) #blue
```

```python
bang=countries.get_group('Bangladesh')
bang_tem=bang.copy()
bang_tem["Life Expectancy"]=(bang_tem["Life Expectancy"]/bang_tem["Life Expectancy"].abs().max())
plt.plot(bang_tem["Life Expectancy"],bang_tem["Year"]) #orange
plt.plot(IND_temp["Life Expectancy"],IND_temp["Year"]) #blue
plt.xlabel('life expectancy')
plt.ylabel('years')
plt.title("LIFE EXPECTANCY IN BANGLADESH AND INDIA")
plt.legend(['bangladesh','india'])
```

```python
# we have to do this with a couple other values as well like beer consumption
bang=countries.get_group('Bangladesh')
bang_temp=bang.copy()
IND_temp["Population"]=(IND_temp["Population in millions"]/(IND_temp["Population in millions"].max()))
bang_temp["Population"]=(bang_temp["Population in millions"]/bang_temp["Population in millions"].abs().max())
plt.plot(bang_temp["Population in millions"],bang_temp["Year"]) #orange
plt.plot(IND_temp["Population in millions"],IND_temp["Year"]) #blue
plt.xlabel('Population in millions')
plt.ylabel('years')
plt.title("POPULATION TRENDS IN BANGLADESH AND INDIA")
plt.legend(['bangladesh','india'])
```
